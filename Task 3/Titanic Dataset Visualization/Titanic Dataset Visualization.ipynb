```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\Ananya\Desktop\DJS Compute Tasks\Task 3\Titanic Dataset Visualization\Titanic Dataset.csv")
df
```

```python
# DETAILS ABOUT THE DATASET
# Survived => Survival => 0 = No, 1 = Yes
# Pclass => Ticket class => 1 = 1st, 2 = 2nd, 3 = 3rd
# Sex => Sex => male or female
# Age => Age in years
# SibSp => no. of siblings / spouses aboard the Titanic
# Parch => no. of parents / children aboard the Titanic
# Ticket => Ticket number
# Fare => Passenger fare
# Cabin => Cabin number
# Embarked => Port of Embarkation => C = Cherbourg, Q = Queenstown, S = Southampton
```

```python
df = df[df['Age'].notnull()]
df = df[df['Embarked'].notnull()]
df = df.drop(columns=['Cabin'])
df
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df = df.reset_index(drop = True)
df
```

```python
df.describe()
```

```python
Pclass_distribution = df['Pclass'].value_counts()
Pclass_distribution
```

```python
Pclass_distribution.plot(kind='bar')
plt.title('Number of Passengers Per Class')
plt.ylabel('Number of Passenger')
plt.xlabel('Passenger Class')
plt.show()
```

```python
Age_distribution = df['Age'].value_counts()
Age_distribution
```

```python
plt.hist(df['Age'], bins = 8)
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
sns.displot(data = df, x = 'Age', bins=[10, 20, 30, 40, 50, 60, 70, 80], hue = 'Survived', multiple='dodge').set(title = 'Age and Survival Status')
plt.show()
```

```python
Sex_distribution = df['Sex'].value_counts()
Sex_distribution
```

```python
plt.pie(Sex_distribution, labels = ['Male', 'Female'])
plt.title('Sex Distribution')
plt.show()
```

```python
sns.countplot(data = df, x = 'Survived', hue = 'Sex').set(title = 'Survival Status and Sex')
plt.show()
```

```python
sns.countplot(data=df, x = 'Pclass', hue = 'Sex').set(title = 'Passenger Class and Sex')
plt.show()
```

```python
plt.boxplot(df['Fare'])
plt.title('Distribution of Fare')
plt.ylabel('Fare')
plt.show()
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, fmt='.1f')
plt.show()
```

```python
sns.violinplot(data = df, x = 'Pclass', y = 'Age').set(title = 'Distribution of Age by Passenger Class')
plt.show()
```

```python
sns.countplot(data = df, x = 'Survived', hue = 'Pclass').set(title='Distribution of Survivors differentiated by Passenger Class')
plt.show()
```

```python
sns.pairplot(data = df, hue = 'Survived')
plt.show()
```
